```python
# demo 1 - demo on scikit learn 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
 
iris = sns.load_dataset('iris')
type(iris)
```

```python
rng = np.random.RandomState(42)
x = 10*rng.rand(50)
print(x.shape)
y = 2*x -1 + rng.rand(50)
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
    
```

```python
x_input = x.reshape(1,-1).T
x_input.shape
```

```python
y_input = y.reshape(1,-1).T
y_input.shape
```

```python
model.fit(x_input,y_input)
print(model.coef_)
print(model.intercept_)
```

```python
xfit = np.linspace(1,10).reshape(1,-1).T
y_predict = model.predict(xfit)
plt.scatter(x,y)
plt.plot(xfit,y_predict)
```
